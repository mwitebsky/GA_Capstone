```python
import pandas as pd
from urllib2 import urlopen
from bs4 import BeautifulSoup
from datetime import date, datetime, timedelta
from time import sleep
```

## Gathers Average Temps and Precip for given day

```python
url = 'https://www.wunderground.com/history/airport/{}/{}/{}/{}/DailyHistory.html'
```

```python
def date_range(start, end):
    dates = []
    curr = start
    while curr <= end:
        dates.append(curr)
        curr += timedelta(days=1)
    return dates

airports = ['KMDW', 'KORD']
start_date = date(2015, 1, 1)
end_date = date(2015, 12, 31)
dates = date_range(start_date, end_date)


data = {}
row = 0
for airport in airports:
    for day in dates:
        print '{} {}-{}-{}'.format(airport, day.year, day.month, day.day)
        soup = BeautifulSoup(urlopen(url.format(airport, day.year, day.month, day.day)))
        lst = []
        data[row] = {} 
        for table in soup('table', {'id' : 'historyTable'}):
            for tr in table('tr'):
                for td in tr('td'):
                    for value in td('span', {'class' : 'wx-value'}):
                        lst.append(value.text)
        data[row]['Airport'] = airport
        data[row]['Date'] = day
        data[row]['Avg_Mean_Temp'] = lst[1]
        data[row]['Avg_Max_Temp'] = lst[3]
        data[row]['Avg_Min_Temp'] = lst[6]
        data[row]['Avg_Precipitation'] = lst[10]
        row += 1
```

```python
df = pd.DataFrame(data).T
```
