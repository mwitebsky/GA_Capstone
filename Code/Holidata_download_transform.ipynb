```python
import os
import pandas as pd
from urllib import urlretrieve
```

```python
years = range(2011, 2021)
for year in years:
    urlretrieve('http://holidata.net/en-US/{}.csv'.format(year), 
                '../Assets/Datasets/Holidata/{}.csv'.format(year))
```

```python
# locate all csv files in directory
directory = '../Assets/Datasets/Holidata//'

file_names = []
for filename in os.listdir(directory):
    if filename.endswith(".csv"): 
        file_names.append(os.path.join(directory, filename))
```

```python
# combine all csvs in pandas
dataframes = []
ct = 1
files = len(file_names)
for csv in file_names:
    dataframes.append(pd.read_csv(csv))
    print 'Added file {} of {}'.format(ct, files)
    ct += 1

df = pd.concat(dataframes)
```

```python
df.drop(['<!DOCTYPE HTML PUBLIC "-//IETF//DTD HTML 2.0//EN">', 'locale', 'notes'], axis=1, inplace=True)
```

```python
df = df[df['region'].isnull() == True]
```

```python
df = df.dropna(subset=['date'])
```
