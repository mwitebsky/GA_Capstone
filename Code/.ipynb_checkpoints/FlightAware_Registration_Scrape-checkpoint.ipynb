```python
import pandas as pd
import numpy as np
from urllib2 import urlopen
from bs4 import BeautifulSoup
from datetime import date, datetime, timedelta
from time import sleep
```

```python
# Get list of all tail numbers
df = pd.read_csv('../Assets/Datasets/BTS_flight_data/Processed/chicago_departures.csv')
tails = df['TAIL_NUM'].unique()
tails = tails.tolist()
```

```python
url = 'https://flightaware.com/resources/registration/{}'
```

```python
data = {}
no_data = []
row = 0
total = len(tails)
for tail in tails:
    print '{} / {}'.format(row, total)
    try:  # if record exists
        soup = BeautifulSoup(urlopen(url.format(tail)))
        data[tail] = {}
        info = soup('div', {'class' : 'medium-3 columns'})[0].text.split('\n')
        plane = info[0].split(' ')
        if (len(plane[0]) == 4) and plane[0].isnumeric():
            data[tail]['mfr_year'] = plane[0]
            start = 1
        else:   # missing year
            data[tail]['mfr_year'] = None
            start = 0
        
        data[tail]['mfr'] = []
        data[tail]['model'] = []
        for item in plane[start:]:
            if (item.isalpha()) & (len(item) > 3) or item == 'INC':
                data[tail]['mfr'].append(item)
            else:
                data[tail]['model'].append(item)
        data[tail]['mfr'] = ' '.join(data[tail]['mfr'])
        data[tail]['model'] = ' '.join(data[tail]['model'])
        specs = info[2].split('/')
        seats = ''
        for num in specs[0]:
            if num.isnumeric():
                seats += str(num)
        data[tail]['seats'] = seats
    except:  # if record doesn't exist
        no_data.append(tail)
    row += 1
data = pd.DataFrame(data).T
```

```python
# clean data
data['N-NUMBER'] = data.index
data = data[data['mfr'].isnull() == False]
data.to_csv('../Assets/Datasets/FlightTracker_Airplane/FA_NNumber_Scrape.csv')
```
